```python
import pandas as pd

# Update this path if you are using a different sampling or data type
input_path = '../../000_data/amex/original_100pct'

# 1. Load the index dataframes
print("Loading data...")
train_idx_df = pd.read_feather(f'{input_path}/df_nn_series_idx_train.feather')
test_idx_df = pd.read_feather(f'{input_path}/df_nn_series_idx_test.feather')
```

```python
def get_seq_len_distribution(df):
    # The feather files contain [i1, i2, customer_ID]
    # Calculate the number of months: i2 - i1 + 1
    seq_lengths = df.iloc[:, 1].astype(int) - df.iloc[:, 0].astype(int) + 1
    df['seq_len'] = seq_lengths
    
    # The 3rd column (index 2) is the customer_ID
    customer_col = df.columns[2]
    
    # Group by the sequence length and count unique customers
    distribution = df.groupby('seq_len')[customer_col].nunique().reset_index()
    distribution.columns = ['Months of Records (seq_len)', 'Number of Distinct Users']
    
    # Calculate percentage for better context
    total_users = distribution['Number of Distinct Users'].sum()
    distribution['Percentage'] = (distribution['Number of Distinct Users'] / total_users * 100).round(2).astype(str) + '%'
    
    return distribution.sort_values('Months of Records (seq_len)')
```

```python

# 2. Process and print Train distribution
print("\n=== TRAIN DATASET DISTRIBUTION ===")
train_dist = get_seq_len_distribution(train_idx_df)
print(train_dist.to_string(index=False))
print(f"Total Train Users: {train_dist['Number of Distinct Users'].sum():,}")
```

```python
# 3. Process and print Test distribution
print("\n=== TEST DATASET DISTRIBUTION ===")
test_dist = get_seq_len_distribution(test_idx_df)
print(test_dist.to_string(index=False))
print(f"Total Test Users: {test_dist['Number of Distinct Users'].sum():,}")
```
